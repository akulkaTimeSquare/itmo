# Importing

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error
from catboost import CatBoostRegressor
from sklearn.compose import TransformedTargetRegressor
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression
import xgboost as xgb
import lightgbm as lgb
import torch
from sklearn.neighbors import KNeighborsRegressor
```

# data preprocessing

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your pandas DataFrame
sns.histplot(data=data, x='price', kde=True)
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
low, high = data.price.quantile([0.01, 0.9])
data = data[(data.price >= low) & (data.price <= high)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your pandas DataFrame
sns.histplot(data=data, x='price', kde=True)
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
numeric = ["kitchen_area", "bath_area", "other_area", "extra_area", "extra_area_count", "year", "ceil_height", "floor_max", "floor", "total_area", "bath_count", "rooms_count"]
categorical = ["gas", "hot_water", "central_heating", "extra_area_type_name", "district_name"]
```

```python
data.drop(columns=['index'], inplace=True)
data.head()
```

```python
data = data.dropna()
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(data[numeric + ["price"]].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.show()
```

```python
to_remove_columns = []
real_numeric = [s for s in numeric if s not in to_remove_columns]
```

```python
data.drop(columns=to_remove_columns, inplace=True)
```

```python
# 1.   (   !)
data['living_area'] = data['total_area'] - data['kitchen_area'] - data['bath_area'] - data['other_area']

# 2.         (  = /)
data['kitchen_share'] = data['kitchen_area'] / data['total_area']

# 3.   (   1-  )
data['floor_norm'] = data['floor'] / data['floor_max'].replace(0, 1)
data['is_first_or_last'] = ((data['floor'] == 1) | (data['floor'] == data['floor_max'])).astype(int)
```

```python
#  
numeric += ['living_area', 'kitchen_share', 'floor_norm']

#  
categorical += ['is_first_or_last']
```

```python
# === 1.   ===
data['age_of_property'] = 2025 - data['year']

# === 2.   ===
data['living_share'] = data['living_area'] / data['total_area']
data['rooms_density'] = data['rooms_count'] / data['total_area']
data['bath_density'] = data['bath_count'] / data['total_area']

# === 3.   ===
data['is_middle_floor'] = ((data['floor'] > 1) & (data['floor'] < data['floor_max'])).astype(int)

# === 4. - ===
for c in ['total_area', 'kitchen_area', 'living_area']:
    data[f'log_{c}'] = np.log1p(data[c])

# === 5.  ===
data['area_x_floor'] = data['total_area'] * data['floor_norm']
data['area_x_rooms'] = data['total_area'] * data['rooms_count']
data['kitchen_x_rooms'] = data['kitchen_area'] * data['rooms_count']

# === 6.    ===
from sklearn.preprocessing import OneHotEncoder
from sklearn.cluster import KMeans

enc = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_dist = enc.fit_transform(data[['district_name']])

kmeans_dist = KMeans(n_clusters=10, random_state=42)
data['district_cluster'] = kmeans_dist.fit_predict(X_dist)

# === 7.   ===
X_layout = data[['total_area','kitchen_area','rooms_count','bath_count']].fillna(0)
kmeans_layout = KMeans(n_clusters=12, random_state=42)
data['layout_cluster'] = kmeans_layout.fit_predict(X_layout)

# ===    ===
numeric += [
    'age_of_property', 'living_share',
    'rooms_density', 'bath_density',
    'area_x_floor', 'area_x_rooms', 'kitchen_x_rooms',
    'log_total_area', 'log_kitchen_area', 'log_living_area'
]

categorical += [
    'is_middle_floor',
    'district_cluster', 'layout_cluster',
]
```

```python
# sns.pairplot(data[numeric])
```

```python
log_vars = [
    "other_area", "floor", "floor_max"
]
```

```python
for var in log_vars:
    data[var] = np.log1p(data[var])
```

```python
from sklearn.preprocessing import LabelEncoder


for var in categorical:
    le = LabelEncoder()
    data[var] = le.fit_transform(data[var])
```

```python
data.head()
```

# Ensemble

```python
import numpy as np
import xgboost as xgb
from catboost import CatBoostRegressor
from sklearn.ensemble import StackingRegressor, ExtraTreesRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.pipeline import Pipeline

# ----    ----
y_train_log = np.log1p(y_train)

cat_best = CatBoostRegressor(
    task_type="GPU", devices="0", loss_function="RMSE",
    n_estimators=1_250, verbose=0, **best_cat,
    random_state=42
)

xgb_best = xgb.XGBRegressor(
    device="cuda", eval_metric="rmse", tree_method="hist",
    n_estimators=5_000, verbosity=0, **best_xgb,
    random_state=42
)

lgd_best = lgb.LGBMRegressor(
    n_estimators=20_000, **best_lgb,
    random_state=42,
)

ridge = Ridge(alpha=1.0)

knn = KNeighborsRegressor(n_neighbors=10, weights="distance")

extra = ExtraTreesRegressor(
    n_estimators=500,
    max_depth=None,
    min_samples_split=4,
    random_state=42
)

estimators = [
    ('catboost', cat_best),
    ('xgboost', xgb_best),
    ('lightgbm', lgd_best),
    ('knn', knn),
    ('ridge', ridge),
    ('extra', extra)
]

stacking_model = StackingRegressor(
    estimators=estimators,
    final_estimator=LinearRegression(),
    cv=5,
    n_jobs=1,
    passthrough=True
)

pipeline = Pipeline([
    ("stacking", stacking_model)
])

pipeline.fit(X_train, y_train_log)
```

```python
pred_log = pipeline.predict(X_test)
pred = np.expm1(pred_log)
```

```python
root_mean_squared_error(y_test, pred)
```
